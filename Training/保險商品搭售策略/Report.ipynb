# 保險商品搭售策略

## 引入函式庫

```python
import pandas as pd
```

## 讀取資料

```python
original_data = pd.read_csv('sales_data.csv', sep=',')
original_data.head()
```

## 資料檢視

```python
cleaned_data = original_data.copy()
```

```python
cleaned_data.columns = ['商品代號', '商品名稱', '銷售通路', '商品分類', '商品單價(單位保費)', '銷售來源', '顧客ID', '顧客性別', '顧客投保年齡', '投保日期與時間', '保單號碼', '總保費', '商品成本']
```

```python
cleaned_data.head()
```

```python
cleaned_data.info()
```

```python
cleaned_data['商品代號'].value_counts()
```

```python
cleaned_data.groupby('商品名稱')[['商品單價(單位保費)', '商品成本']].mean()
```

```python
cleaned_data['商品利潤'] = cleaned_data['商品單價(單位保費)'] - cleaned_data['商品成本']
```

```python
cleaned_data.head()
```

```python
top3_type_df = cleaned_data.groupby('商品分類')[['總保費']].sum()
top3_type_df = top3_type_df.sort_values(by='總保費', ascending=False)
top3_type_df = top3_type_df.head(3)
top3_type_df
```

```python
top3_product_df = cleaned_data.groupby(['商品代號', '商品名稱', '商品分類'])[['總保費']].sum()
top3_product_df = top3_product_df.sort_values(by='總保費', ascending=False)
top3_product_df = top3_product_df.head(3)
top3_product_df
```
